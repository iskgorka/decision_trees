```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Documents/Go Data Science/titanic_train.csv")
data.head()
```

```python
round(data['Fare'].median(),2)
```

```python
round(data['Fare'].std(),2)
```

```python
# np.inf - бесконечность
#round(data.groupby(pd.cut(data.Age, [0, 30, 60, np.inf]))['Survived'].mean().mul(100),1)
round(100 * data[data['Age'] < 30]['Survived'].mean(),1),round(100 * data[data['Age'] > 60]['Survived'].mean(),1)
```

```python
round(100*data[data['Sex'] == 'male']['Survived'].mean(),1),round(100*data[data['Sex'] == 'female']['Survived'].mean(),1)
```

```python
data[data['Sex'] == 'male']['Name'].apply(lambda full_name: full_name.split(',')[1].split()[1]).value_counts().head()
```

```python
data[data['Survived'] == 1]['Fare'].hist(color="green", 
                                         label='Survived')
data[data['Survived'] == 0]['Fare'].hist(color="red", 
                                         label='Died')
plt.title('Ticket fare for survived and died')
plt.xlabel('Pounds')
plt.ylabel('Frequency')
plt.legend();
    
```

```python
data[data['Survived'] == 1]['Age'].hist(color="green", 
                                         label='Survived', alpha=.5)
data[data['Survived'] == 0]['Age'].hist(color="red", 
                                         label='Died', alpha=.5)
plt.title('Age for survived and died')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.legend();
```

```python
import seaborn as sns
sns.boxplot(data['Survived'], data['Age']);
data.groupby('Survived')['Age'].mean()
```
