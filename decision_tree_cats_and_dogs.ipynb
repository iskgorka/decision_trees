```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
from IPython.display import HTML
style = "<style>svg{width: 120% !important; height: 120% !important;} </style>"
HTML(style)
```

```python
data = pd.read_csv('https://stepik.org/media/attachments/course/4852/dogs_n_cats.csv')
```

```python
data.head(2)
```

```python
X = data.drop(['Вид'], axis=1)
y = data['Вид']
```

```python
clf = tree.DecisionTreeClassifier(class_weight=None,criterion='entropy',
                                 max_depth=5, max_features=None)
```

```python
clf.fit(X, y)
```

```python
from graphviz import Source
from IPython.display import SVG
```

```python
graph = Source(tree.export_graphviz(clf, out_file=None,
                                   feature_names=list(X), class_names=['собачка','котик'],
                                   filled = True))
display(SVG(graph.pipe(format='svg')))
```
